```python
def ejecutar_sql(query, conexion=conn):
  """
  Toma una consulta SQL, la ejecuta y devuelve el resultado
  en una tabla de pandas para una visualización clara.
  """
  df = pd.read_sql_query(query, conexion)
  # Usamos display() porque en Colab presenta las tablas de forma más elegante.
  display(df)
```

```python
# Total secuestrados

ejemplo = """
    SELECT SUM(cantidad) AS total_secuestrados
    FROM secuestrados
"""
ejecutar_sql(ejemplo)
```

```python
# total de secuestros por municipio

ejemplo1 = """
    SELECT municipio, SUM(cantidad) AS total_secuestrados
    FROM secuestrados
    GROUP BY municipio
    ORDER BY total_secuestrados DESC
"""
ejecutar_sql(ejemplo1)
```

```python
# total de secuestros por tipo de delito y departamento

ejemplo2 = """
    SELECT departamento, tipo_delito, SUM(cantidad) AS total_secuestrados
    FROM secuestrados
    GROUP BY departamento, tipo_delito
    ORDER BY total_secuestrados DESC
"""
ejecutar_sql(ejemplo2)
```
